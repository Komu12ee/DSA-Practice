```python
import pandas as pd
```

```python
FILE_PATH = "./district.txt"
FILE_PATH_UT = "./district_ut.txt"
```

```python
state = []
district = []
with open(FILE_PATH_UT, 'r') as file:
    data = file.readlines()
    for line in data:
        line = line.strip()
        if len(line) == 0:
            continue
        if len(line) <= 30:
            state.append(line)
        else:
            district.append(line.split(', '))

lines = []
for s, d in zip(state, district):
    for dist in d:
        print(f"{s},{dist}")
        lines.append(f"{s},{dist}\n")
```
